```python
import os
from ultralytics import YOLO
from roboflow_dataset import RoboflowDataset
```

```python
RF_VER_NUM = 2 #check that this version exists in ur dir first
RF_PROJECT = "fingerprint-pattern-classifier"
```

```python
parent_dir = os.path.abspath(os.path.join(os.getcwd(), "../.."))
dataset_dir = os.path.abspath(os.path.join(parent_dir, f"datasets/{RF_PROJECT}/v{RF_VER_NUM}/"))
print(dataset_dir)
```

```python
if not os.path.exists(dataset_dir):
    rf_dataset = RoboflowDataset(rf_project=RF_PROJECT, rf_dataset_ver=RF_VER_NUM)
    rf_dataset.download_dataset(export_type="folder", 
                                dataset_dir=os.path.abspath(dataset_dir)
                                )
else:
    print('ok')
```

```python
model = YOLO("yolov8s-cls.pt")
```

```python
model.train(
    data=f"{dataset_dir}/original_dataset",  # auto set by roboflow
    epochs=20,
    imgsz=256,
    batch=32,
    dropout = 0.2,
    save_period = 1,
    patience = 5,
    val = True,
    name="fingerprint-classification"
)
```

```python
test_image_dir = os.path.abspath(os.path.join(os.getcwd(), "test_images/"))
model = YOLO("/Users/jin/Documents/GitHub/AI-Project/runs/classify/fingerprint-classification5/weights/best.pt")
for pattern in os.listdir(test_image_dir):
    if pattern != ".DS_Store":
        results = model.predict(source=f"test_images/{pattern}", save=True)
        print(results)
```
