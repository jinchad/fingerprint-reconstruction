```python
import os
from roboflow import Roboflow
from ultralytics import YOLO
```

```python
RF_KEY  = os.environ.get("RF_KEY", "")
RF_WORKSPACE = os.environ.get("RF_WORKSPACE", "")
RF_PROJECT = os.environ.get("RF_CLASSIFICATION_PROJECT", "")
RF_CLASSIFIER_VER_NUM = os.environ.get("RF_CLASSIFIER_VER_NUM", None)
print(RF_KEY, RF_WORKSPACE, RF_PROJECT, RF_CLASSIFIER_VER_NUM)
```

```python
rf = Roboflow(api_key=RF_KEY)  # üîÅ Replace with your actual API key
project = rf.workspace(RF_WORKSPACE).project(RF_PROJECT)
dataset = project.version(RF_CLASSIFIER_VER_NUM).download("folder", f"../../datasets/Fingerprint classification/Fingerprint classification dataset v{RF_CLASSIFIER_VER_NUM}")
```

```python
model = YOLO("yolov8s-cls.pt")
```

```python
model.train(
    data=f"Fingerprint classification/Fingerprint classification dataset v{RF_CLASSIFIER_VER_NUM}",  # auto set by roboflow
    epochs=7,
    imgsz=512,
    batch=64,
    dropout = 0.3,
    save_period = 1,
    val = True,
    name="fingerprint-classification"
)
```
