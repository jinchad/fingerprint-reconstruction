```python
import os
from roboflow import Roboflow
from ultralytics import YOLO
```

```python
RF_KEY  = os.environ.get("RF_KEY", "")
RF_WORKSPACE = os.environ.get("RF_WORKSPACE", "")
RF_PROJECT = os.environ.get("RF_CLASSIFICATION_PROJECT", "")
VERSION_NUM = 4
print(RF_KEY, RF_WORKSPACE, RF_PROJECT)
```

```python
rf = Roboflow(api_key=RF_KEY)  # üîÅ Replace with your actual API key
project = rf.workspace(RF_WORKSPACE).project(RF_PROJECT)
dataset = project.version(VERSION_NUM).download("folder", f"../../datasets/Fingerprint classification dataset v{VERSION_NUM}")
```

```python
model = YOLO("yolov8s-cls.pt")
```

```python
model.train(
    data=f"Fingerprint classification dataset v{VERSION_NUM}",  # auto set by roboflow
    epochs=5,
    imgsz=512,
    batch=128,
    dropout = 0.3,
    save_period = 1,
    val = True,
    name="fingerprint-classification"
)
```

```python
results = model.predict(source="test images/", save=True)
print(results)
```
