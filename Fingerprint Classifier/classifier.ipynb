```python
import os
from ultralytics import YOLO
from roboflow_dataset import RoboflowDataset
```

```python
RF_VER_NUM = 4 #check that this version exists in ur dir first
RF_PROJECT = "fingerprint-classification-un8qd"
```

```python
model = YOLO("yolov8s-cls.pt")
```

```python
parent_dir = os.path.abspath(os.path.join(os.getcwd(), "../.."))
dataset_dir = os.path.abspath(os.path.join(parent_dir, f"datasets/{RF_PROJECT}/v{RF_VER_NUM}/original_dataset/"))
print(dataset_dir)
```

```python
if not os.path.exists(dataset_dir):
    rf_dataset = RoboflowDataset(rf_project=RF_PROJECT, rf_dataset_ver=4)
    rf_dataset.download_dataset(export_type="folder", 
                                dataset_dir=os.path.abspath(dataset_dir)
                                )
else:
    print('ok')
```

```python
model.train(
    data=dataset_dir,  # auto set by roboflow
    epochs=7,
    imgsz=512,
    batch=32,
    dropout = 0.3,
    save_period = 1,
    val = True,
    name="fingerprint-classification"
)
```

```python
results = model.predict(source="test images/", save=True)
print(results)
```
